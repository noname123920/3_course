## –ó–∞–¥–∞–Ω–∏–µ 4: –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø—Ä–æ–≤–µ—Å—Ç–∏ –∞—Ç–∞–∫—É –Ω–∞ –∑–∞–∫—Ä—ã—Ç—ã–π –∫–ª—é—á, —Ç. –µ. –ø–æ–¥–æ–±—Ä–∞—Ç—å f(x) –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ h(x).



## –ê—Ç–∞–∫–∞ –Ω–∞ –∑–∞–∫—Ä—ã—Ç—ã–π –∫–ª—é—á –≤ –∫–æ–ª—å—Ü–µ \( R = \mathbb{Z}_2[x]/(x^5+1) \)

## –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
- **–ö–æ–ª—å—Ü–æ**: $R = \mathbb{Z}_2[x]/(x^5+1)$
- **–û—Ç–∫—Ä—ã—Ç—ã–π –∫–ª—é—á**: $h(x) = x^4 + x^2 + x$
- **–ó–∞–∫—Ä—ã—Ç—ã–π –∫–ª—é—á**: $f(x) = x^3 + x + 1$

### –®–∞–≥ 1: –ê–Ω–∞–ª–∏–∑ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –∫–æ–ª—å—Ü–∞

–ö–æ–ª—å—Ü–æ \( R = \mathbb{Z}_2[x]/(x^5+1) \) –∏–º–µ–µ—Ç –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç—å: –ø–æ–ª–∏–Ω–æ–º \( x^5+1 \) —Ä–∞—Å–∫–ª–∞–¥—ã–≤–∞–µ—Ç—Å—è –Ω–∞–¥ \( \mathbb{Z}_2 \):

$$
x^5 + 1 = (x + 1)(x^4 + x^3 + x^2 + x + 1)
$$

–≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –∫–æ–ª—å—Ü–æ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –ø–æ–ª–µ–º –∏ —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–µ–ª–∏—Ç–µ–ª–∏ –Ω—É–ª—è.

### –®–∞–≥ 2: –ü–æ–∏—Å–∫ —Å–≤—è–∑–∏ –º–µ–∂–¥—É –∫–ª—é—á–∞–º–∏

–í NTRU-–ø–æ–¥–æ–±–Ω—ã—Ö —Å–∏—Å—Ç–µ–º–∞—Ö –æ–±—ã—á–Ω–æ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ:
$$ f(x) \cdot h(x) \equiv g(x) \pmod{q} $$

–ù–æ –≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ \( q = 2 \), –∏ –º—ã —Ä–∞–±–æ—Ç–∞–µ–º –≤ \( \mathbb{Z}_2 \). –ü–æ–ø—Ä–æ–±—É–µ–º –Ω–∞–π—Ç–∏ –ø—Ä–æ—Å—Ç—É—é —Å–≤—è–∑—å.

–£–º–Ω–æ–∂–∏–º –æ—Ç–∫—Ä—ã—Ç—ã–π –∫–ª—é—á –Ω–∞ —Ä–∞–∑–ª–∏—á–Ω—ã–µ –ø—Ä–æ—Å—Ç—ã–µ –ø–æ–ª–∏–Ω–æ–º—ã:

**–ü—Ä–æ–±—É–µ–º —É–º–Ω–æ–∂–∏—Ç—å –Ω–∞ \( x \):**
$$
\begin{align*}
x \cdot h(x) &= x \cdot (x^4 + x^2 + x) \\
&= x^5 + x^3 + x^2 \\
&\equiv 1 + x^3 + x^2 \pmod{x^5+1}
\end{align*}
$$

**–ü—Ä–æ–±—É–µ–º —É–º–Ω–æ–∂–∏—Ç—å –Ω–∞ \( x^2 \):**
$$
\begin{align*}

&= x^6 + x^4 + x^3 \\
&\equiv x + x^4 + x^3 \pmod{x^5+1}
\end{align*}
$$

**–ü—Ä–æ–±—É–µ–º —É–º–Ω–æ–∂–∏—Ç—å –Ω–∞ \( x^3 \):**
$$
\begin{align*}
x^3 \cdot h(x) &= x^3 \cdot (x^4 + x^2 + x) \\
&= x^7 + x^5 + x^4 \\
&\equiv x^2 + 1 + x^4 \pmod{x^5+1}
\end{align*}
$$

**–ü—Ä–æ–±—É–µ–º —É–º–Ω–æ–∂–∏—Ç—å –Ω–∞ \( x^4 \):**
$$
\begin{align*}

&= x^8 + x^6 + x^5 \\
&\equiv x^3 + x + 1 \pmod{x^5+1}
\end{align*}
$$

### –®–∞–≥ 3: –û–±–Ω–∞—Ä—É–∂–µ–Ω–∏–µ –∑–∞–∫—Ä—ã—Ç–æ–≥–æ –∫–ª—é—á–∞

–°—Ä–∞–≤–Ω–∏–≤–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –Ω–∞ \( x^4 \) —Å –∏—Å–∫–æ–º—ã–º –∑–∞–∫—Ä—ã—Ç—ã–º –∫–ª—é—á–æ–º:

$$ x^4 \cdot h(x) = x^3 + x + 1 = f(x) $$

**–ü–æ–ª—É—á–∞–µ–º**: 
$$ f(x) = x^4 \cdot h(x) $$

### –®–∞–≥ 4: –ü—Ä–æ–≤–µ—Ä–∫–∞

–ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ –Ω–∞–π–¥–µ–Ω–Ω—ã–π –∫–ª—é—á –∫–æ—Ä—Ä–µ–∫—Ç–µ–Ω:

$$
\begin{align*}
f(x) \cdot h(x) &= (x^3 + x + 1)(x^4 + x^2 + x) \\
&= x^7 + x^5 + x^4 + x^6 + x^4 + x^3 + x^4 + x^2 + x \\
&= x^7 + x^6 + x^5 + x^4 + x^3 + x^2 + x
\end{align*}
$$

–ü—Ä–∏–≤–æ–¥–∏–º –ø–æ –º–æ–¥—É–ª—é \( x^5+1 \):
$$
\begin{align*}
&\equiv x^2 + x + 1 + x^4 + x^3 + x^2 + x \pmod{x^5+1} \\
&= x^4 + x^3 + 2x^2 + 2x + 1 \\
&= x^4 + x^3 + 1 \quad \text{(–≤ } \mathbb{Z}_2\text{)}
\end{align*}

```python
def poly_mod(poly, mod_poly):
    """
    –†–µ–¥—É–∫—Ü–∏—è –ø–æ–ª–∏–Ω–æ–º–∞ –ø–æ –º–æ–¥—É–ª—é x^5 + 1
    –ò—Å–ø–æ–ª—å–∑—É–µ–º —Å–≤–æ–π—Å—Ç–≤–æ: x^5 ‚â° 1, x^6 ‚â° x, x^7 ‚â° x^2, –∏ —Ç.–¥.
    """
    result = [0] * 5
    for power, coeff in enumerate(poly):
        reduced_power = power % 5
        result[reduced_power] = (result[reduced_power] + coeff) % 2
    return result

def poly_mult(a, b):
    """
    –£–º–Ω–æ–∂–µ–Ω–∏–µ –ø–æ–ª–∏–Ω–æ–º–æ–≤ –≤ Z2[x]
    """
    max_degree = len(a) + len(b) - 1
    result = [0] * max_degree
    
    for i, coeff_a in enumerate(a):
        if coeff_a == 1:
            for j, coeff_b in enumerate(b):
                if coeff_b == 1:
                    result[i + j] = (result[i + j] + 1) % 2
    return result

def poly_add(a, b):
    """
    –°–ª–æ–∂–µ–Ω–∏–µ –ø–æ–ª–∏–Ω–æ–º–æ–≤ –≤ Z2[x]
    """
    max_len = max(len(a), len(b))
    result = [0] * max_len
    
    for i in range(max_len):
        coeff_a = a[i] if i < len(a) else 0
        coeff_b = b[i] if i < len(b) else 0
        result[i] = (coeff_a + coeff_b) % 2
    return result

def poly_to_string(poly):
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø–æ–ª–∏–Ω–æ–º–∞ –≤ —Å—Ç—Ä–æ–∫—É
    """
    terms = []
    for power, coeff in enumerate(poly):
        if coeff == 1:
            if power == 0:
                terms.append("1")
            elif power == 1:
                terms.append("x")
            else:
                terms.append(f"x^{power}")
    
    if not terms:
        return "0"
    
    return " + ".join(terms)

# –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
print("=== –®–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ —Å–æ–æ–±—â–µ–Ω–∏—è –≤ –∫–æ–ª—å—Ü–µ Z2[x]/(x^5+1) ===")

# –ü–æ–ª–∏–Ω–æ–º—ã –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ–º –∫–∞–∫ —Å–ø–∏—Å–∫–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –æ—Ç –º–ª–∞–¥—à–µ–π —Å—Ç–µ–ø–µ–Ω–∏ –∫ —Å—Ç–∞—Ä—à–µ–π
# –ù–∞–ø—Ä–∏–º–µ—Ä: x^4 + x^2 + x = [0, 1, 1, 0, 1]

h = [0, 1, 1, 0, 1]  # h(x) = x^4 + x^2 + x
r = [0, 1, 0, 1, 0]  # r(x) = x^3 + x  
m = [1, 0, 1, 0, 0]  # m(x) = x^2 + 1

print(f"–û—Ç–∫—Ä—ã—Ç—ã–π –∫–ª—é—á h(x) = {poly_to_string(h)}")
print(f"–°–ª—É—á–∞–π–Ω—ã–π –ø–æ–ª–∏–Ω–æ–º r(x) = {poly_to_string(r)}")
print(f"–°–æ–æ–±—â–µ–Ω–∏–µ m(x) = {poly_to_string(m)}")

# –®–∞–≥ 1: –í—ã—á–∏—Å–ª—è–µ–º r(x) * h(x)
print("\n--- –®–∞–≥ 1: –í—ã—á–∏—Å–ª—è–µ–º r(x) * h(x) ---")
r_times_h = poly_mult(r, h)
print(f"r(x) * h(x) = {poly_to_string(r_times_h)}")

# –†–µ–¥—É–∫—Ü–∏—è –ø–æ –º–æ–¥—É–ª—é x^5 + 1
r_times_h_mod = poly_mod(r_times_h, [1, 0, 0, 0, 0, 1])  # x^5 + 1
print(f"r(x) * h(x) mod (x^5+1) = {poly_to_string(r_times_h_mod)}")

# –®–∞–≥ 2: –°–∫–ª–∞–¥—ã–≤–∞–µ–º —Å m(x)
print("\n--- –®–∞–≥ 2: –°–∫–ª–∞–¥—ã–≤–∞–µ–º —Å —Å–æ–æ–±—â–µ–Ω–∏–µ–º m(x) ---")
e = poly_add(r_times_h_mod, m)
print(f"e(x) = r(x)*h(x) + m(x) = {poly_to_string(e)}")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä—É—á–Ω–æ–≥–æ —Ä–∞—Å—á–µ—Ç–∞
print("\n--- –ü—Ä–æ–≤–µ—Ä–∫–∞ ---")
expected_e = [1, 0, 1, 1, 1]  # x^4 + x^3 + x^2 + 1
print(f"–û–∂–∏–¥–∞–µ–º—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç: {poly_to_string(expected_e)}")
print(f"–§–∞–∫—Ç–∏—á–µ—Å–∫–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç: {poly_to_string(e)}")
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ–≤–ø–∞–¥–∞—é—Ç: {e == expected_e}")

print(f"\nüéØ –ó–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ: e(x) = {poly_to_string(e)}")
```

```python
def poly_mod(poly, mod_poly):
    """
    –†–µ–¥—É–∫—Ü–∏—è –ø–æ–ª–∏–Ω–æ–º–∞ –ø–æ –º–æ–¥—É–ª—é x^5 + 1
    """
    result = [0] * 5
    for power, coeff in enumerate(poly):
        reduced_power = power % 5
        result[reduced_power] = (result[reduced_power] + coeff) % 2
    return result

def poly_mult(a, b):
    """
    –£–º–Ω–æ–∂–µ–Ω–∏–µ –ø–æ–ª–∏–Ω–æ–º–æ–≤ –≤ Z2[x]
    """
    max_degree = len(a) + len(b) - 1
    result = [0] * max_degree
    
    for i, coeff_a in enumerate(a):
        if coeff_a == 1:
            for j, coeff_b in enumerate(b):
                if coeff_b == 1:
                    result[i + j] = (result[i + j] + 1) % 2
    return result

def poly_add(a, b):
    """
    –°–ª–æ–∂–µ–Ω–∏–µ –ø–æ–ª–∏–Ω–æ–º–æ–≤ –≤ Z2[x] (–≤ Z2 –≤—ã—á–∏—Ç–∞–Ω–∏–µ = —Å–ª–æ–∂–µ–Ω–∏–µ)
    """
    max_len = max(len(a), len(b))
    result = [0] * max_len
    
    for i in range(max_len):
        coeff_a = a[i] if i < len(a) else 0
        coeff_b = b[i] if i < len(b) else 0
        result[i] = (coeff_a + coeff_b) % 2
    return result

def poly_to_string(poly):
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø–æ–ª–∏–Ω–æ–º–∞ –≤ —Å—Ç—Ä–æ–∫—É
    """
    terms = []
    for power, coeff in enumerate(poly):
        if coeff == 1:
            if power == 0:
                terms.append("1")
            elif power == 1:
                terms.append("x")
            else:
                terms.append(f"x^{power}")
    
    if not terms:
        return "0"
    
    return " + ".join(terms)

# –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è —Ä–∞—Å—à–∏—Ñ—Ä–æ–≤–∫–∏
print("=== –†–∞—Å—à–∏—Ñ—Ä–æ–≤–∫–∞ —Å–æ–æ–±—â–µ–Ω–∏—è –≤ –∫–æ–ª—å—Ü–µ Z2[x]/(x^5+1) ===")

# –ü–æ–ª–∏–Ω–æ–º—ã –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ–º –∫–∞–∫ —Å–ø–∏—Å–∫–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –æ—Ç –º–ª–∞–¥—à–µ–π —Å—Ç–µ–ø–µ–Ω–∏ –∫ —Å—Ç–∞—Ä—à–µ–π
e = [1, 0, 1, 1, 1]      # e(x) = x^4 + x^3 + x^2 + 1
r = [0, 1, 0, 1, 0]      # r(x) = x^3 + x
h = [0, 1, 1, 0, 1]      # h(x) = x^4 + x^2 + x
expected_m = [1, 0, 1, 0, 0]  # –û–∂–∏–¥–∞–µ–º–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ: x^2 + 1

print(f"–®–∏—Ñ—Ä—Ç–µ–∫—Å—Ç e(x) = {poly_to_string(e)}")
print(f"–°–ª—É—á–∞–π–Ω—ã–π –ø–æ–ª–∏–Ω–æ–º r(x) = {poly_to_string(r)}")
print(f"–û—Ç–∫—Ä—ã—Ç—ã–π –∫–ª—é—á h(x) = {poly_to_string(h)}")
print(f"–û–∂–∏–¥–∞–µ–º–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ m(x) = {poly_to_string(expected_m)}")

# –§–æ—Ä–º—É–ª–∞ —Ä–∞—Å—à–∏—Ñ—Ä–æ–≤–∫–∏: m(x) = e(x) + r(x) * h(x) mod (x^5+1)
print("\n--- –®–∞–≥ 1: –í—ã—á–∏—Å–ª—è–µ–º r(x) * h(x) ---")
r_times_h = poly_mult(r, h)
print(f"r(x) * h(x) = {poly_to_string(r_times_h)}")

# –†–µ–¥—É–∫—Ü–∏—è –ø–æ –º–æ–¥—É–ª—é x^5 + 1
r_times_h_mod = poly_mod(r_times_h, [1, 0, 0, 0, 0, 1])
print(f"r(x) * h(x) mod (x^5+1) = {poly_to_string(r_times_h_mod)}")

print("\n--- –®–∞–≥ 2: –í—ã—á–∏—Å–ª—è–µ–º m(x) = e(x) + r(x)*h(x) ---")
decrypted_m = poly_add(e, r_times_h_mod)
print(f"m(x) = e(x) + r(x)*h(x) = {poly_to_string(decrypted_m)}")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print("\n--- –ü—Ä–æ–≤–µ—Ä–∫–∞ ---")
print(f"–†–∞—Å—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ: {poly_to_string(decrypted_m)}")
print(f"–û–∂–∏–¥–∞–µ–º–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ: {poly_to_string(expected_m)}")
print(f"–†–∞—Å—à–∏—Ñ—Ä–æ–≤–∫–∞ —É—Å–ø–µ—à–Ω–∞: {decrypted_m == expected_m}")

# –î–µ—Ç–∞–ª—å–Ω–æ–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å–ª–æ–∂–µ–Ω–∏—è
print("\n--- –î–µ—Ç–∞–ª—å–Ω–æ–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ ---")
print("e(x) = x^4 + x^3 + x^2 + 1")
print("r(x)*h(x) = x^4 + x^3")
print("m(x) = (x^4 + x^3 + x^2 + 1) + (x^4 + x^3)")
print("–í Z2: x^4 + x^4 = 0, x^3 + x^3 = 0")
print("m(x) = x^2 + 1 ‚úÖ")
```
