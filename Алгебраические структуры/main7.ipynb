### 4. –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ

## –®–∞–≥ 1: –ó–∞–ø–∏—à–µ–º —É—Å–ª–æ–≤–∏–µ —Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏

–ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, —á—Ç–æ:

–≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç:
$$ (a \cdot x_1 + b)\mod{p} = (a \cdot x_2 + b)\mod{p}
$$
–ò–ª–∏ —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ:
$$ a \cdot x_1 + b \equiv a \cdot x_2 + b \pmod{p}
$$

## –®–∞–≥ 2: –£–ø—Ä–æ—Å—Ç–∏–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ

–í—ã—á–∏—Ç–∞–µ–º $b$ –∏–∑ –æ–±–µ–∏—Ö —á–∞—Å—Ç–µ–π —Å—Ä–∞–≤–Ω–µ–Ω–∏—è:
$$ a \cdot x_1 \equiv a \cdot x_2 \pmod{p}
$$
–¢–µ–ø–µ—Ä—å –º–æ–∂–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç—å –æ–±–µ —á–∞—Å—Ç–∏ –Ω–∞ $a$, —Ç–∞–∫ –∫–∞–∫ $a \neq 0$. –ù–æ –ø–µ—Ä–µ–¥ —ç—Ç–∏–º –≤–∞–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å –æ–¥–Ω–æ —Å–≤–æ–π—Å—Ç–≤–æ –ø–æ–ª–µ–π –∫–æ–Ω–µ—á–Ω—ã—Ö —á–∏—Å–µ–ª $\mathbb{GF}(p)$: –ª—é–±–æ–µ –Ω–µ–Ω—É–ª–µ–≤–æ–µ —á–∏—Å–ª–æ –∏–º–µ–µ—Ç –æ–±—Ä–∞—Ç–Ω–æ–µ –ø–æ —É–º–Ω–æ–∂–µ–Ω–∏—é. –¢–∞–∫ –∫–∞–∫ $a \neq 0$, —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ç–∞–∫–æ–µ —á–∏—Å–ª–æ $a^{-1}$, —á—Ç–æ:
$$ a \cdot a^{-1} \equiv 1 \pmod{p}
$$
–£–º–Ω–æ–∂–∏–≤ –æ–±–µ —Å—Ç–æ—Ä–æ–Ω—ã –Ω–∞—à–µ–≥–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –Ω–∞ $a^{-1}$ –ø–æ–ª—É—á–∞–µ–º:
$$ x_1 \equiv x_2 \pmod{p}

## –®–∞–≥ 3: –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞

–¢–∞–∫ –∫–∞–∫ –º—ã –ø–æ–ª—É—á–∏–ª–∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ $ x_1 \equiv x_2 \pmod{p} $, —ç—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ $x_1$ –∏ $x_2$ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ –Ω–∞ –≤–µ–ª–∏—á–∏–Ω—É, –∫—Ä–∞—Ç–Ω—É—é $p$. –û–¥–Ω–∞–∫–æ –ø–æ—Å–∫–æ–ª—å–∫—É –º—ã —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –Ω–∞ –º–Ω–æ–∂–µ—Å—Ç–≤–µ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª $\{0, 1, ..., p-1\}$, —Ä–∞–∑–Ω–∏—Ü–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å —Ä–∞–≤–Ω–∞ –Ω—É–ª—é, –∏–Ω–∞—á–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤—ã–π–¥—É—Ç –∑–∞ –ø—Ä–µ–¥–µ–ª—ã –¥–∏–∞–ø–∞–∑–æ–Ω–∞.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–º —Ä–µ—à–µ–Ω–∏–µ–º —è–≤–ª—è–µ—Ç—Å—è:
$$ x_1 = x_2
$$

## –ó–∞–∫–ª—é—á–µ–Ω–∏–µ

–ú—ã –¥–æ–∫–∞–∑–∞–ª–∏, —á—Ç–æ –µ—Å–ª–∏ $ h(x_1) = h(x_2) $, —Ç–æ $ x_1 = x_2 $. –≠—Ç–æ –∏ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –Ω–∞—à–∞ —Ñ—É–Ω–∫—Ü–∏—è $ h(x) $ —è–≤–ª—è–µ—Ç—Å—è –∏–Ω—ä–µ–∫—Ü–∏–µ–π. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –µ—Å–ª–∏ $ a \neq 0 $, —Ö–µ—à-—Ñ—É–Ω–∫—Ü–∏—è –≥–∞—Ä–∞–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ —É–Ω–∏–∫–∞–ª—å–Ω–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤—Ö–æ–¥–∞ –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö –º–Ω–æ–∂–µ—Å—Ç–≤–∞ $\mathbb{Z}_p$.

---

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –¥–æ–∫–∞–∑–∞–Ω–æ, —á—Ç–æ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ $ a \neq 0 $ –¥–∞–Ω–Ω–∞—è —Ö–µ—à-—Ñ—É–Ω–∫—Ü–∏—è –∏–Ω—ä–µ–∫—Ç–∏–≤–Ω–∞.

```python
def hash_function(x, a, b, p):
    return (a * x + b) % p

def check_collisions(p):
    import random
    a = random.randint(1, p-1)
    b = random.randint(0, p-1)
    seen_hashes = set()
    
    for x in range(p):
        hashed_val = hash_function(x, a, b, p)
        if hashed_val in seen_hashes:
            return f"–ö–æ–ª–ª–∏–∑–∏—è –¥–ª—è p={p}, a={a}, b={b}"
        seen_hashes.add(hashed_val)
    
    return f"–ù–µ—Ç –∫–æ–ª–ª–∏–∑–∏–π –¥–ª—è p={p}, a={a}, b={b}"

# –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ
primes = [7, 11, 13, 17, 19]
for prime in primes:
    print(check_collisions(prime))
```
